```python
SalesHistory = SalesHistory.merge(CustomerMasterData, on='Customer ID', how='left') \
.merge(LocationMasterData, on='Location ID', how='left') \
.merge(ProductMasterData, on='Product ID', how='left')
SalesHistory.head()
```

```python
SalesHistory = SalesHistory.drop(columns=["Customer ID", "Location ID", "Product ID"])
SalesHistory.head()
```

```python
df = SalesHistory.copy()
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
# Latitude
df['Geo Latitude'] = (
    df['Geo Latitude']
    .str.replace('°', '')
    .str.replace(' ', '')
)

df['Geo Latitude'] = df['Geo Latitude'].apply(
    lambda x: -float(x.replace('S', '')) if 'S' in x else float(x.replace('N', ''))
)

# Longitude
df['Geo Longitude'] = (
    df['Geo Longitude']
    .str.replace('°', '')
    .str.replace(' ', '')
)

df['Geo Longitude'] = df['Geo Longitude'].apply(
    lambda x: -float(x.replace('W', '')) if 'W' in x else float(x.replace('E', ''))
)

df['Geo Latitude'] = pd.to_numeric(df['Geo Latitude'], errors='coerce')
df['Geo Longitude'] = pd.to_numeric(df['Geo Longitude'], errors='coerce')
```
